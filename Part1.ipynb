```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data_cleaned = pd.read_csv('C:/Users/wang/Desktop/NewFolder/data_cleaned.csv')

sns.set(style="whitegrid")
```

```python
for column in data_cleaned.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(data_cleaned[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.savefig(f'C:/Users/wang/Desktop/NewFolder/{column}_distribution.png')
    plt.show()
```

```python
for column in data_cleaned.select_dtypes(include=['object']).columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(y=column, data=data_cleaned, order = data_cleaned[column].value_counts().index)
    plt.title(f'Distribution of {column}')
    plt.xlabel('Count')
    plt.ylabel(column)
    plt.savefig(f'C:/Users/wang/Desktop/NewFolder/{column}_countplot.png')
    plt.show()
```
